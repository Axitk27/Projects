```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
with open('tablet.html',encoding ='utf-8') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html,'lxml')
soup.prettify()
```

```python
# Gathering Information 


names = []
prices = []
ratings = []
info1 = []
info2 = []
info3 = []
info4 = []
info5 = []
info6 = []
info7 = []
info8 = []

for i in soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}):
    try:
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
        prices.append(i.find('span',{'class':'price'}).text)
    except:
        prices.append(np.nan)
    try:
        ratings.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        ratings.append(np.nan)
        
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    
    
    
    try:
        info1.append(x[0].text)
    except:
        info1.append(np.nan)
    try:
        info2.append(x[1].text)
    except:
        info2.append(np.nan)
    try:    
        info3.append(x[2].text)
    except:
        info3.append(np.nan)
    try:
        info4.append(x[3].text)
    except:
        info4.append(np.nan)
    try:
        info5.append(x[4].text)
    except:
        info5.append(np.nan)
    try:
        info6.append(x[5].text)
    except:
        info6.append(np.nan)
        
    try:
        info7.append(x[6].text)
    except:
        info7.append(np.nan)
        
    try:
        info8.append(x[7].text)
    except:
        info8.append(np.nan)
```

```python

df = pd.DataFrame({
    'model':names,
    'price':prices,
    'rating':ratings,
    'info1':info1,
    'info2':info2,
    'info3':info3,
    'info4':info4,
    'info5':info5,
    'info6':info6,
    'info7':info7,
    'info8':info8
})
```
